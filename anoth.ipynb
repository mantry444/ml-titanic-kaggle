```python
p=pd.get_dummies(file, columns=['Sex', 'Embarked'])
file1=p
file1
```

```python
from sklearn.ensemble import RandomForestClassifier 
features= ['Pclass','SibSp','Parch','Fare','Sex_female','Sex_male','Embarked_C', 'Embarked_S','Embarked_Q']
X=file1[features]
X
```

```python
Y=file1['Survived']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.2)
model= RandomForestClassifier(n_estimators=1000,max_depth=10,n_jobs=-1 )
model.fit(x_train, y_train)
```

```python
model.base_estimator_
```

```python
predict=model.predict(x_test)
```

```python
train_score=model.score(x_train,y_train)
train_score
```

```python
from  sklearn.metrics import accuracy_score
acuracy= accuracy_score(y_test, predict)
acuracy
```

```python
%%timeit
model= RandomForestClassifier(n_jobs=-1)
model.fit(x_train,y_train)
```

```python
%%timeit
model= RandomForestClassifier(n_jobs=1)
model.fit(x_train,y_train)
```
